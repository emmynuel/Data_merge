## The report are seperated and many  
1. 1st open the directory of files in loop
1. Merge those report together into a new csv file
1. Save the merge report as new and read the first few lines to confirm

```python
# First open the folder

files = [file for file in os.listdir('./downloads/Now-MTN')]

all_data = pd.DataFrame()

# Create a code to loop through the files in the folder

for file in files:
   df = pd.read_csv("./Downloads/Now-MTN/"+file)

# Join and create a new joined csv file 
    
all_data = pd.concat([df for file in files])

all_data.to_csv("./Downloads/Now-MTN/COMBINE.csv", index=False)
  
all_data
```


#### Check numbers of roles and columns

```python
# check the number of rows and column of the new form file

df = pd.read_csv("./downloads/Mtn_report/here.csv")
df.shape
```

#### Delete the empty columns

```python
# removing an empty column

column_to_delete = ["Initiated By","On Behalf Of","From Handler Name","Amount","Currency","External Amount","Currency.1","External FX Rate","External Service Provider","Provider Category","Currency.2","Information","Note/Message","Discount","Currency.3","Promotion","Currency.4","Coupon","Currency.5","Balance","Currency.6"]
df = df.drop(columns = column_to_delete)
```

#### Create a new column for month and day to allow us analyse our sales monthly

```python
# Creating a month and day column to enable us analyse

df['Month'] = df['Date'].str[5:7]
df['Day'] = df['Date'].str[8:10]
df
```

### Check if there are unsuccessful Transactions 

```python
check = df['Status'] != "Successful"
check.nunique()
```

#### Analyse the report by summing up the fee generated per day

```python
sumUp = df.groupby('Day').agg({'Fee': 'sum'}).round()
sumUp.style.highlight_min()
```
